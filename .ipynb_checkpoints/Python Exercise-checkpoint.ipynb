```python
#3.
def fun_recursion(n):
    if n == 1:
        return n
    else:
        return n * fun_recursion(n-1)
    
num = int(input("Enter a non negative integer"))

if num < 0:
    print("The factorial does not exists for a negative number.")
elif num == 0:
    print("The factorial of a number {} is {}".format(num,1))
else:
    print("The factorial of the number {} is {}".format(num, fun_recursion(num)))
```

```python
#9.
def fun_duplicate(s):
    b = ""
    for i in s:
        c = 0
        for j in s:
            if(i == j):
                c = c + 1
            else:
                pass
        if c == 1:
            b = b + i
        else:
            pass
    return b 

st = input("Enter a string ")
fun_duplicate(st)
```
